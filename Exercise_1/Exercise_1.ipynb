### Question 8
   Consider the following code snippet
   
        m, c = 2 , -4
        N = 10
        x = np . linspace (0 , N-1, N) . reshape (N, 1 )
        sigma = 10
        y = m*x + c + np . random . normal (0 , sigma , (N, 1 ))

   (a) Append a column of ones to x to create X.<br>
   (b) Compute [X<sup>⊺</sup>X]<sup>−1</sup>X<sup>⊺</sup>y

```python
m, c = 2 ,-4
N = 10
x = np.linspace(0,N-1,N).reshape(N,1)
sigma = 10
y = m*x + c + np.random.normal(0,sigma,(N,1))

X = np.append(x,np.ones((N,1)),axis=1)
print(np.dot(np.dot(np.linalg.inv(np.dot(X.T,X)),X.T),y))
```

### Question 10
Use Gaussian smoothing to filter the noise in the image gal_gaussian.png

```python
import cv2 as cv

Image=cv.imread("gal_gaussian.png")
cv.namedWindow("Image",cv.WINDOW_AUTOSIZE)
cv.imshow("Image",Image)

Blur_Image1=cv.GaussianBlur(Image,(5,5),0)
cv.namedWindow("Blurred_Image1",cv.WINDOW_AUTOSIZE)
cv.imshow("Blurred_Image1",Blur_Image1)

Blur_Image2=cv.GaussianBlur(Image,(25,25),0)
cv.namedWindow("Blurred_Image2",cv.WINDOW_AUTOSIZE)
cv.imshow("Blurred_Image2",Blur_Image2)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

figure, axis = plt.subplots(1,3)
figure.set_figwidth(20)

axis[0].imshow(cv.cvtColor(Image, cv.COLOR_BGR2RGB))
axis[1].imshow(cv.cvtColor(Blur_Image1, cv.COLOR_BGR2RGB))
axis[2].imshow(cv.cvtColor(Blur_Image2, cv.COLOR_BGR2RGB))
plt.show()
```

### Question 11
Use median filtering to filter the salt and pepper noise in the image gal_sandp.png

```python
import cv2 as cv

Image=cv.imread("gal_sandp.png")
cv.namedWindow("Image",cv.WINDOW_AUTOSIZE)
cv.imshow("Image",Image)

Blur_Image1=cv.medianBlur(Image,5)
cv.namedWindow("Blurred_Image1",cv.WINDOW_AUTOSIZE)
cv.imshow("Blurred_Image1",Blur_Image1)

Blur_Image2=cv.medianBlur(Image,25)
cv.namedWindow("Blurred_Image2",cv.WINDOW_AUTOSIZE)
cv.imshow("Blurred_Image2",Blur_Image2)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

figure, axis = plt.subplots(1,3)
figure.set_figwidth(20)

axis[0].imshow(cv.cvtColor(Image, cv.COLOR_BGR2RGB))
axis[1].imshow(cv.cvtColor(Blur_Image1, cv.COLOR_BGR2RGB))
axis[2].imshow(cv.cvtColor(Blur_Image2, cv.COLOR_BGR2RGB))
plt.show()
```

### Question 14
Increase the brightness of the image tom_dark.jpg

```python
import cv2 as cv

Image=cv.imread("tom_dark.jpg")
cv.namedWindow("Image",cv.WINDOW_AUTOSIZE)
cv.imshow("Image",Image)

Bright_Im=Image+30
cv.namedWindow("Bright_Im",cv.WINDOW_AUTOSIZE)
cv.imshow("Bright_Im",Bright_Im)

cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

figure, axis = plt.subplots(1,2)
figure.set_figwidth(15)

axis[0].imshow(cv.cvtColor(Image, cv.COLOR_BGR2RGB))
axis[1].imshow(cv.cvtColor(Bright_Im, cv.COLOR_BGR2RGB))
plt.show()
```
