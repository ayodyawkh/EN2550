### Question 1
Carry out gamma correction on the image show in Fig. 1 with γ ∈ [0.2, 0.8, 1.2, 2]

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread('spider.png')
assert f is not None

cv.imshow('Image',f)
cv.waitKey(0)

gamma_list = [0.2,0.8,1.2,2]
for gamma in gamma_list:
    t = np.array([(p/255)**gamma*255 for p in range (0,256)]).astype(np.uint8) 
    g = cv.LUT(f,t)
    figure, axis = plt.subplots()
    axis.plot(t)
    axis.set_aspect('equal')
    
    cv.namedWindow("Image",cv.WINDOW_AUTOSIZE)
    cv.imshow('Image',g)
    cv.waitKey(0)
    cv.destroyAllWindows()
```

```python
def gamma_coordinate(f,gamma):
    t = np.array([(p/255)**gamma*255 for p in range (0,256)]).astype(np.uint8) 
    g = cv.LUT(f,t) 
    return g

figure, axis = plt.subplots(1,5)
figure.set_figwidth(20)

axis[0].imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
for i,gamma in enumerate(gamma_list):
    g = gamma_coordinate(f,gamma)
    axis[i+1].imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
plt.show()
```

### Question 2
Implement the intensity transformation in Fig. 2 on the image shown in Fig. 1

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread('spider.png')
assert f is not None
t1 = np.linspace(0,50,100)
t2 = np.linspace(50,200,156)

t = np.concatenate((t1,t2),axis=0).astype(np.uint8)
figure, axis = plt.subplots()
axis.plot(t)
axis.set_aspect('equal')
assert len(t) == 256
g = cv.LUT(f,t)

cv.namedWindow("Image",cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
figure, axis = plt.subplots(1,2)
figure.set_figwidth(15)

axis[0].imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
axis[1].imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
plt.show()
```

### Question 3
Fig. 3 shows a somewhat dark image. <br>
(a) Show the histogram of the original image. <br>
(b) Histogram equalize the image. <br>
(c) Show the histogram of the equalized image.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread('shells.tif', cv.IMREAD_GRAYSCALE)
assert f is not None

hist_f = cv.calcHist([f],[0], None, [256], [0,256])
g = cv.equalizeHist(f)
hist_g = cv.calcHist([g],[0], None, [256], [0,256])

figure, axis = plt.subplots(2,1, figsize = (8,8))
axis[0].plot(hist_f)
axis[1].plot(hist_g)

cv.namedWindow("Image",cv.WINDOW_AUTOSIZE)
cv.imshow('Image before',f)
cv.waitKey(0)
cv.imshow('Image after',g)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
figure, axis = plt.subplots(1,2)
figure.set_figwidth(10)

axis[0].imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
axis[1].imshow(cv.cvtColor(g, cv.COLOR_BGR2RGB))
plt.show()
```

### Question 4
Fig. 4 shows an image probably of the Zion park. <br>
(a) Enhance the saturation of this image. <br>
(b) Change the hue to your liking.

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f = cv.imread('zion_pass.jpg')
assert f is not None

hsv_Img = cv.cvtColor(f, cv.COLOR_BGR2HSV)
h,s,v = cv.split(hsv_Img)

s = cv.add(s , 40)

hsv_img = cv.merge([h,s,v])
Img = cv.cvtColor(hsv_img, cv.COLOR_HSV2BGR)

cv.namedWindow("Image",cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',Img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
figure, axis = plt.subplots(1,2)
figure.set_figwidth(15)

axis[0].imshow(cv.cvtColor(f, cv.COLOR_BGR2RGB))
axis[1].imshow(cv.cvtColor(Img, cv.COLOR_BGR2RGB))
plt.show()
```
